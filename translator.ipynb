Translating csv file in english to spanish, german and french

```python
from googletrans import Translator


def transspanish(word):
 translations = translator.translate(word, dest="es")
 for translation in translations:
    res= (f"{translation.text}")
    return res


def transfrench(word):
 translations = translator.translate(word, dest="fr")
 for translation in translations:
    res= (f"{translation.text}")
    return res

def transgerman(word):
 translations = translator.translate(word, dest="de")
 for translation in translations:
    res= (f"{translation.text}")
    return res

import csv
  
# Open file 
with open('/content/English.csv') as file_obj:
      
    # Create reader object by passing the file 
    # object to reader method
 reader_obj = csv.reader(file_obj)
      
    # Iterate over each row in the csv 
    # file using reader object
 print('text in spanish')
 for row in reader_obj:
        translated_text= transspanish(row)   
        print(translated_text)

 print('text in french')
 for row in reader_obj:
        translated_text= transfrench(row)   
        print(translated_text)

 print('text in german')
 for row in reader_obj:
        translated_text= transgerman(row)   
        print(translated_text)              
```
